## Merge data from each month into one CSV

```python
import pandas as pd

# Assuming df1, df2, ..., df12 are your DataFrames
df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]

# Concatenate all DataFrames in the list
df = pd.concat(df_list, ignore_index=True)

# Display the first few rows of the concatenated DataFrame
df.head()
```

```python
df.shape
```

## Clean-up the Data



# 1) Drop NaN rows

```python
import pandas as pd



# Create a boolean mask where True indicates the presence of NaN values
nan_mask = df.isna()

# Use the boolean mask to filter the DataFrame and display rows where NaN values exist
rows_with_nan = df[nan_mask.any(axis=1)]

# Display the rows where NaN values exist
rows_with_nan
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
df.head()
```

## Find "OR" data and delete it

```python
df = df[df['Order Date'].str[0:2]!='Or']
```

## Make columns correct type

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])
```

# Coverts colun=ms to correct type

```python
df["Quantity Ordered"]=pd.to_numeric(df["Quantity Ordered"]) # Make it to int
df["Price Each"]=pd.to_numeric(df["Price Each"]) # Make it to float
```

## Adding data with additional columns

## 1) Add Month Column

```python
df['Month'] = df['Order Date'].str[0:2]
df['Month'] = df['Month'].astype('int32')
df.head()
```

## 2) Add Sales column

```python
df["Sales"]=df["Quantity Ordered"] * df["Price Each"]
df.head()
```

## 3) Add a city column

```python
# Create a function to get city
def Get_city(address):
  return address.split(",")[1]

# Create a function to get State
def Get_State(address):
  return address.split(",")[2].split(" ")[1]


df["City"]=df["Purchase Address"].apply(lambda x : f"{Get_city(x)} ({Get_State(x)})")
df.head()
# df["City"] = df["Purchase Address"].apply(lambda x: x.split(",")[1] + " " + x.split(",")[2].split(" ")[1])
# df.head()
```

## Data Exploration!

#### Question 1 : What was the best month for sales?How much was earned that month?

```python
import pandas as pd

# Group data by 'Month' and calculate total sales for each month
monthly_sales = df.groupby('Month')['Sales'].sum().reset_index()

# Display the result
print(monthly_sales)
```

```python
import pandas as pd

# Assuming df is your DataFrame with the provided data

# Group data by 'Month' and calculate total sales for each month
monthly_sales_million = df.groupby('Month')['Sales'].sum().reset_index()

# Convert sales to millions
monthly_sales_million['Sales (Million)'] = monthly_sales_million['Sales'] / 1000000

# Display the result
print(monthly_sales_million)
```

```python
import matplotlib.pyplot as plt
plt.bar(monthly_sales_million["Month"],monthly_sales_million["Sales"])
plt.xlabel("Months")
plt.ylabel("Sales in Million")
plt.show()
```

```python
import plotly.graph_objs as go


# Create a bar trace with custom color
trace = go.Bar(x=monthly_sales["Month"], y=monthly_sales["Sales"], marker=dict(color='green'))

# Define the layout
layout = go.Layout(title="Monthly Sales",
                   xaxis=dict(title="Months"),
                   yaxis=dict(title="Sales in USD ($)"))

# Create a Figure object
fig = go.Figure(data=[trace], layout=layout)

# Show the chart
fig.show()

```

#### Q 3: What time should we display the advertisement to maximize the likehood of customers buying products

```python
# Convert Order Date in Datetime format
df["Order Date"]=pd.to_datetime(df["Order Date"])
df.head()
```

#### Create two column Hour and Minute

```python
df["Hour"]=df["Order Date"].dt.hour
df["Minute"]=df['Order Date'].dt.minute
df.head()
```

```python
# Counting the number of order ids for each hour
highest_Sales_Hour = df.groupby('Hour')['Order ID'].count().reset_index()

# Display the result
print(highest_Sales_Hour)
```

```python
import plotly.graph_objs as go

# Create a line trace
trace = go.Scatter(x=highest_Sales_Hour["Hour"],
                   y=highest_Sales_Hour["Order ID"],
                   mode='lines',
                   name='Line',
                   line=dict(color='black'))

# Create a layout
layout = go.Layout(title='Hourly Order Count',
                   xaxis=dict(title="Hours"),
                   yaxis=dict(title="Count of Orders"))

# Create a Figure object
fig = go.Figure(data=[trace], layout=layout)

# Show the chart
fig.show()
```

#### Q4: What Products sold the most?Why it is sold the most?

```python
# Grouping by "Product" and summing the "Quantity Ordered"
product_group = df.groupby("Product")
Quantity_Ordered = product_group["Quantity Ordered"].sum().reset_index()

print(Quantity_Ordered)
```

```python
import plotly.graph_objs as go


# Create a bar trace
trace = go.Bar(x=Quantity_Ordered["Product"],  # Use the "Product" column as x-axis values
               y=Quantity_Ordered["Quantity Ordered"],
               marker=dict(color='darkolivegreen'))  # You can customize the color if needed

# Create a layout
layout = go.Layout(title="Quantity Ordered for each Product",
                   xaxis=dict(title="Products"),
                   yaxis=dict(title="Quantity Ordered"))

# Create a Figure object
fig = go.Figure(data=[trace], layout=layout)

# Show the chart
fig.show()
```
